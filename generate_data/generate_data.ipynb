```python
# Bedrock Quotas - https://docs.aws.amazon.com/bedrock/latest/userguide/quotas.html
token_max = 200000
request_max = 200
inference_sec = 1 # time (sec) per single inference

# Inference parameters
max_tokens = 10000
temperature = 0.0
top_p = 0.7
top_k = 50

#############################################
### Start inference
#############################################

model_type = model_id.split('.')[0]
model_name = model_id.split('.')[1].split(':')[0]
```

```python
client = None
def initialize():
  global client
  client = boto3.client(service_name='bedrock-runtime')
```

```python
def invoke_anthropic(prompt):
    body = {
            "anthropic_version": "bedrock-2023-05-31",
            "max_tokens": max_tokens,
            "temperature": temperature,
            "top_p": top_p,
            "top_k": top_k,
            "messages": [{
                         "role": "user",
                         "content": [{"type": "text", "text": prompt}],
                        }]
            }
    
    response = client.invoke_model(modelId=model_id, body=json.dumps(body))

    response_body = json.loads(response.get("body").read())
    output_list = response_body.get("content", [])

    response = output_list[0]["text"]
        
    return response
```
